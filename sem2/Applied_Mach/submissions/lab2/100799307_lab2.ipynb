### Import libraries

```python
# Import libraries required to train the model and predict the images
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.applications.inception_v3 import decode_predictions
```

### Loading the Inception V3 network

```python
# Transfer learning using a pre-trained InceptionV3 network
model = InceptionV3()
#Print model summary to understand whats happening in the backend
model.summary()
```

### Processing the image to be added to the model

```python
# Loading the image named sample_image.jpg and setting the size as 299,299 as it is required for InceptionV3 model
image = load_img('sample_image.jpg', target_size = (299,299))
# Converting image to array
image = img_to_array(image)
# Reshaping the single image
image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
# Preprocess function preprocesses the image for the pre-trained model
image = preprocess_input(image)
```

### The image that we are going to use as an input to our model




### Predicting the class of the image

```python
# Classify the image 
yhat = model.predict(image)
# This will list down all the possible classes for the image with probability
label = decode_predictions(yhat)
label
```

```python
# Finally, we can list down the class with the highest probability
label[0][0]
```

```python
print("Class of the image is : {}" .format(label[0][0][1]))
print("With probability: {}%" .format(label[0][0][2]*100))
```

### Here, the Inception V3 network classified the image as "African Elephant" with a probability of 79.25%
