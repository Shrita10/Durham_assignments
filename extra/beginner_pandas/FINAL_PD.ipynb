```python
#cleaning the location column
import pandas as pd
import numpy as np
df = pd.read_csv('Documents/dataset/anime_subdata.csv')
```

```python
def seg_nos(s):
    s = str(s)
    if s == 'United Kingdom':
        return s
    elif ',' in s:
        s = s.split(',')
        return s[1]
    elif ' ' in s:
        s = s.split()
        return s[1]
    elif (s == ''):
        s = s.replace('','Austria')
        return s
    else:
        return s
df['location'] = df['location'].apply(seg_nos)
```
