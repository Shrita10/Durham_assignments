```python
math['school_new'] = math['school'].map({'GP' : 1, 'MS' : 2})
math['health_new'] = math['health'].map({1: 'bad health', 2: 'bad health', 3: 'medium health', 4: 'Good health', 5: 'Good health'})
math['Walc'] = math['Walc'].map({1 : 'low', 2 : 'low', 3: 'medium', 4:'high', 5:'very high'})
math['Dalc'] = math['Dalc'].map({1 : 'low', 2 : 'low', 3: 'medium', 4:'high', 5:'very high'})
math['traveltime'] = math['traveltime'].map({1: '<15m', 2: '15-30m', 3: '30m-1h', 4: '>1h'})
math['Medu_new'] = math['Medu'].map({0: 'No education', 1 : 'Primary education', 2 : '5-9 Grade education', 3: 'Secondary education', 4:'higher education'})
math['Fedu_new'] = math['Fedu'].map({0: 'No education', 1 : 'Primary education', 2 : '5-9 Grade education', 3: 'Secondary education', 4:'higher education'})
math['studytime_new'] = math['studytime'].map({1 : '<2 hours', 2 : '2-5 hours', 3: '5-10 hours', 4:'> 10 hours'})
math['famrel'] = math['famrel'].map({1 : 'very bad', 2 : 'bad', 3: 'good', 4:'good',5:'excellent'})
math['freetime'] = math['freetime'].map({1 : 'very low', 2 : 'low', 3: 'mdeium', 4:'high',5: 'very high'})
math['goout'] = math['goout'].map({1 : 'very low', 2 : 'low', 3: 'medium', 4:'high',5: 'very high'})
```

```python
port['school_new'] = port['school'].map({'GP' : 1, 'MS' : 2})
port['health_new'] = port['health'].map({1: 'bad health', 2: 'bad health', 3: 'medium health', 4: 'Good health', 5: 'Good health'})
port['Walc'] = port['Walc'].map({1 : 'low', 2 : 'low', 3: 'medium', 4:'high', 5:'very high'})
port['Dalc'] = port['Dalc'].map({1 : 'low', 2 : 'low', 3: 'medium', 4:'high', 5:'very high'})
port['traveltime'] = port['traveltime'].map({1: '<15m', 2: '15-30m', 3: '30m-1h', 4: '>1h'})
port['Medu_new'] = port['Medu'].map({0: 'No education', 1 : 'Primary education', 2 : '5-9 Grade education', 3: 'Secondary education', 4:'higher education'})
port['Fedu_new'] = port['Fedu'].map({0: 'No education', 1 : 'Primary education', 2 : '5-9 Grade education', 3: 'Secondary education', 4:'higher education'})
port['studytime_new'] = port['studytime'].map({1 : '<2 hours', 2 : '2-5 hours', 3: '5-10 hours', 4:'> 10 hours'})
port['famrel'] = port['famrel'].map({1 : 'very bad', 2 : 'bad', 3: 'good', 4:'good',5:'excellent'})
port['freetime'] = port['freetime'].map({1 : 'very low', 2 : 'low', 3: 'mdeium', 4:'high',5: 'very high'})
port['goout'] = port['goout'].map({1 : 'very low', 2 : 'low', 3: 'medium', 4:'high',5: 'very high'})
```
